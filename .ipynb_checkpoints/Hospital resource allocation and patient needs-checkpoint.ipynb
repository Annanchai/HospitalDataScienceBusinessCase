## Abnormality check

```python
#Check for abnormal values in Age group
df['Age'].unique()
```

11-20 age group has been recorded as Nov-20 needs to be changed to 11-20

```python
def groupNameChange(x):
    if x == 'Nov-20':
        return '11-20'
    else:
        return x
```

```python
df['Age'] = df['Age'].apply(groupNameChange)
```

```python
df['Age'].unique()
```

```python
df['Stay'].unique()
```

Again 11-20 age group has been recorded as Nov-20 needs to be changed to 11-20

```python
df['Stay'] = df['Stay'].apply(groupNameChange)
```

```python
df['Stay'].unique()
```

## Univariate analysis

```python
plt.figure(figsize=(12,8))
for i, column in enumerate(numerical_columns.columns, 1):
    plt.subplot(3,3,i)
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
for i, column in enumerate(numerical_columns.columns, 1):
    plt.subplot(3,3,i)
    sns.boxplot(data=df, x=column)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```

## Exploring Patient Demographics

```python
#Age wise analysis
plt.figure(figsize=(8,5))
sns.countplot(data=df, x=df['Age'])
plt.title('Age vs Patient count')
plt.show()
```

```python
ageCount = df['Age'].value_counts()
```

```python
plt.figure(figsize=(8,5))
plt.pie(ageCount, labels=ageCount.index, autopct='%1.1f%%')
plt.title('Age vs Patient count')
plt.show()
```
